Let us generate some random data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

### 1. Data Generation

```python
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X**2 + 2 + np.random.randn(m, 1)
```

Next, let's fit a second-degree polynomial to our X inputs

### 2. Model Development

```python
#Generate polynomial features up to the 2nd degree and don't include a bias column
poly_Features = PolynomialFeatures(degree = 2, include_bias = False)

#Fit the PolynomialFeatures transformer to the input data X and then transforms X into polynomial features
X_poly = poly_Features.fit_transform(X)
```

Now let us fit a linear regression model on the extended training data

```python
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
lin_reg.intercept_, lin_reg.coef_
```

### 3. Visualization

Plot the polynomial Regression model predictions

But first, let's plot the features and predictions

```python
plt.scatter(X, y, color='blue', label='Data points')
```

Next, let us fit the polynomial model to the plot

np.linspace generates 100 evenly spaced values between the minimum and maximum values of X.

This creates a smooth set of points across the range of X to be used for plotting the regression line.
.reshape(-1, 1):

This reshapes the generated values into a column vector (2D array) with one column and an appropriate number of rows.

-1 in the reshape function automatically infers the number of rows based on the length of the array and the specified number of columns (1 in this case).

Uses the trained linear regression model lin_reg to predict the y values corresponding to the transformed polynomial features X_range_poly.

```python
X_range = np.linspace(min(X), max(X), 100).reshape(-1, 1)
X_range_poly = poly_Features.transform(X_range)
y_range_pred = lin_reg.predict(X_range_poly)
plt.plot(X_range, y_range_pred, color='red', label='Polynomial Regression Line')
```

And then combine the two

```python
plt.title('Polynomial Regression Model Predictions')
plt.xlabel('Features (X)')
plt.ylabel('Output/Predictions (y)')
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X_range, y_range_pred, color='red', label='Polynomial Regression Line')
plt.legend()
```
