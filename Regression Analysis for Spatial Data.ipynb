### Multicollinearity

It is the occurrence of high intercorrelation among two or more independent variable in a multiple regression model.

We can scatterplots to test for multicollinearity

```python
fig = plt.figure(figsize = (8, 6))
sns.axes_style('darkgrid')
#Clay vs NDVI
plt.subplot(2, 3, 1)
sns.scatterplot(x = 'clay', y = 'vi', data = gdf, color = 'green')
sns.axes_style('darkgrid')
plt.xlabel('Clay')
plt.ylabel('NDVI')
plt.title('NDVI vs Clay')

#Prcp vs NDVI
plt.subplot(2, 3, 2)
sns.scatterplot(x = 'prcp', y = 'vi', data = gdf, color = 'green')
sns.axes_style('darkgrid')
plt.xlabel('Precipitation (mm)')
plt.ylabel('NDVI')
plt.title('NDVI vs Precipitation')

#Perm vs NDVI
plt.subplot(2, 3, 3)
sns.scatterplot(x = 'perm', y = 'vi', data = gdf, color = 'green')
sns.axes_style('darkgrid')
plt.xlabel('Permeability (inc/hr)')
plt.ylabel('NDVI')
plt.title('NDVI vs Permeability')

#Perm vs Clay
plt.subplot(2, 3, 4)
sns.scatterplot(x = 'perm', y = 'clay', data = gdf, color = 'green')
sns.axes_style('darkgrid')
plt.xlabel('Permeability (inc/hr)')
plt.ylabel('Clay')
plt.title('Clay vs Permeability')

#Clay vs Precipitation
plt.subplot(2, 3, 5)
sns.scatterplot(x = 'prcp', y = 'clay', data = gdf, color = 'green')
sns.axes_style('darkgrid')
plt.xlabel('Precipitation')
plt.ylabel('clay')
plt.title('Clay vs Precipitation')

#perm vs prcp
plt.subplot(2, 3, 6)
sns.scatterplot(x = 'perm', y = 'prcp', data = gdf, color = 'green')
plt.xlabel('Permeability (inc/hr)')
plt.ylabel('Precipitation')
plt.title('Permeability vs Precipitation')

plt.tight_layout()
plt.show()
fig.savefig('multicollinearity.png')
```

### Correlation

We check for correlation of variables by plotting a heatmap

```python
vars = gdf[['vi', 'clay', 'perm', 'prcp']]
corr = vars.corr()
```

```python
corr
```

```python
fig = plt.figure(figsize = (8, 6))
heatmap = sns.heatmap(corr, annot = True, cmap = 'YlGnBu')
plt.tight_layout()
fig.savefig('correlation.png')
plt.show()
```

### Linear Regression

##### 1. NDVI against Clay

```python
y = gdf['vi']
x = gdf['clay']
x = sm.add_constant(x)
model = sm.OLS(y, x)
results = model.fit()
sns.set_style('darkgrid')
plt.scatter(gdf['clay'], gdf['vi'], label = 'Data')
plt.plot(gdf['clay'], results.fittedvalues, color = 'red', label = 'Model')
plt.xlabel('Clay')
plt.ylabel('NDVI')
plt.title('Linear Regression Model')
plt.legend()

# Adjust layout
plt.tight_layout()

# Save the plot as a PNG file
plt.savefig('best_fit.png')

# Show the plot
plt.show()
```

```python
print(results.summary())
```

##### 2. NDVI against Permeability

```python
y = gdf['vi']
x = gdf['perm']
x = sm.add_constant(x)
model = sm.OLS(y, x)
results = model.fit()

plt.scatter(gdf['perm'], gdf['vi'], label = 'Data')
plt.plot(gdf['perm'], results.fittedvalues, color = 'green', label = 'Predicted')

plt.xlabel('Permeability (inc/hr)')
plt.ylabel('NDVI')
plt.legend()
plt.show()
plt.tight_layout()
fig.savefig('linear_regression.png')
```

```python
print(results.summary())
```

#### 3. Multiple Regression

```python
y = gdf['vi']
x = gdf[['clay', 'prcp', 'perm']]
x = sm.add_constant(x)
model = sm.OLS(y, x)
results = model.fit()
print(results.summary())
```
